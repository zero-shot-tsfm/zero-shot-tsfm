# WB

```r
############################
# 1) نصب و لود پکیج
############################
install.packages("urca")
library(urca)
############################
# 2) خواندن CSV
############################
df <- read.csv("/content/WB90.csv")
############################
# 3) ستون‌هایی که نادیده گرفته می‌شوند
############################
ignore_cols <- c("cpcb_time", "row_id")
############################
# 4) دیتافریم خروجی
############################
results <- data.frame(
  series = character(),
  ADF = numeric(),
  stationary_5pct = logical(),
  stringsAsFactors = FALSE
)
############################
# 5) اجرای ADF و ذخیره خروجی
############################
for (col in colnames(df)) {
  if (col %in% ignore_cols) next
  if (!is.numeric(df[[col]])) next
  if (sum(!is.na(df[[col]])) == 0 || var(df[[col]], na.rm = TRUE) == 0) next
  res <- ur.df(
    df[[col]],
    type = "drift",
    lags = 1,
    selectlags = "Fixed"
  )
  # مقدار ADF (tau statistic) با گرد کردن تا دو رقم اعشار
  adf_value <- round(res@teststat[1], 2)
  # مقدار بحرانی ۵٪
  crit_5pct <- res@cval[1, "5pct"]
  # تصمیم ایستایی
  stationary <- adf_value < crit_5pct
  # اضافه به جدول
  results <- rbind(
    results,
    data.frame(
      series = col,
      ADF = adf_value,
      stationary_5pct = stationary
    )
  )
}
############################
# 6) نمایش خروجی
############################
print(results)
```
