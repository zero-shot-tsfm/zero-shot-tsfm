```r
install.packages("scoringRules")
library(scoringRules)

samples <- as.matrix(read.csv("chronos_samples.csv", header = FALSE))
y_true  <- as.numeric(read.csv("y_true.csv", header = FALSE)[,1])

# CRPS برای هر زمان
crps_t <- crps_sample(y = y_true, dat = samples)

# میانگین CRPS
mean_crps <- mean(crps_t)
print(mean_crps)
```

```r
crps_norm_exact <- sum(crps_t) / sum(abs(y_true))
print(crps_norm_exact)
```
