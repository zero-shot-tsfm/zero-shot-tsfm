```python
# -*- coding: utf-8 -*-
# =============================================================
#    Simple TCN-like encoder + dense decoder
#    Multivariate forecasting (open,high,low,close,volume)
#    HISTORY = 249, FORECAST = 63
#    بدون هیچ اسکیلینگ
# =============================================================

!pip install -q tensorflow numpy pandas matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import Input, Model
from tensorflow.keras.layers import Conv1D, BatchNormalization, Activation, Add, Dropout, GlobalAveragePooling1D, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error

# -----------------------------
# 1) mount drive & load csv
# -----------------------------
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv(
    '/content/drive/MyDrive/data/FINAL SELECTED DATA SET/Bitcoin.csv',
    sep=';',
    parse_dates=['merge_date'],
    dayfirst=True
)
print(df.shape)
print(df.head())

# -----------------------------
# 2) select features
# -----------------------------
features = ['open','high','low','close','volume']
data = df[features].astype(float).values
n_features = data.shape[1]

HISTORY = 249
FORECAST = 63

if data.shape[0] < HISTORY + FORECAST:
    raise ValueError("طول سری کمتر از ۲۴۹+۶۳ است!")

# -----------------------------
# 3) prepare training and real future
# -----------------------------
train_data = data[-(HISTORY+FORECAST):-FORECAST]   # 249 آخر برای آموزش
actual_future = data[-FORECAST:]                  # 63 آخر واقعی
X = train_data.reshape(1, HISTORY, n_features)
Y = actual_future.reshape(1, FORECAST, n_features)

# -----------------------------
# 4) TCN residual block
# -----------------------------
def residual_tcn_block(x, filters, kernel_size, dilation_rate, dropout_rate=0.1):
    conv1 = Conv1D(filters, kernel_size, padding='causal', dilation_rate=dilation_rate)(x)
    conv1 = BatchNormalization()(conv1)
    conv1 = Activation('relu')(conv1)
    conv1 = Dropout(dropout_rate)(conv1)
    conv2 = Conv1D(filters, kernel_size, padding='causal', dilation_rate=dilation_rate)(conv1)
    conv2 = BatchNormalization()(conv2)
    res = x
    if x.shape[-1] != filters:
        res = Conv1D(filters, 1, padding='same')(x)
    out = Add()([res, conv2])
    out = Activation('relu')(out)
    return out

# -----------------------------
# 5) build model
# -----------------------------
inp = Input(shape=(HISTORY, n_features))
x = inp

dilations = [1,2,4,8]
filters = 64
kernel_size = 3
for d in dilations:
    x = residual_tcn_block(x, filters=filters, kernel_size=kernel_size, dilation_rate=d)

x = residual_tcn_block(x, filters=128, kernel_size=3, dilation_rate=16)

x = GlobalAveragePooling1D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.2)(x)
out = Dense(FORECAST * n_features, activation='linear')(x)
out_reshaped = tf.keras.layers.Reshape((FORECAST, n_features))(out)

model = Model(inputs=inp, outputs=out_reshaped)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# -----------------------------
# 6) train model
# -----------------------------
history = model.fit(
    X, Y,
    epochs=60,
    batch_size=1,
    verbose=2
)

# -----------------------------
# 7) inference
# -----------------------------
pred = model.predict(X)[0]  # shape (63, n_features)

# -----------------------------
# 8) plotting
# -----------------------------
history_close = train_data[:,3]   # close history
pred_close = pred[:,3]           # close predicted
actual_close = actual_future[:,3]  # close actual

history_idx = np.arange(HISTORY)
future_idx = np.arange(HISTORY, HISTORY+FORECAST)

# full plot
plt.figure(figsize=(16,6))
plt.plot(history_idx, history_close, label='History (249)', linewidth=2, color='orange')
plt.plot(future_idx, actual_close, label='Real Future (63)', linewidth=2, color='green')
plt.plot(future_idx, pred_close, '--', label='Predicted Future (TCN)', linewidth=2, color='red')
plt.title("TCN Forecasting — Close Price (بدون اسکیلینگ)", fontsize=14)
plt.xlabel("Time Index")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# zoom future
plt.figure(figsize=(12,5))
plt.plot(future_idx, actual_close, label='Real Future (close)', linewidth=2, color='green')
plt.plot(future_idx, pred_close, '--', label='Predicted Future (TCN)', linewidth=2, color='red')
plt.title("Zoom: Real vs Predicted — 63 Future Steps (TCN)")
plt.xlabel("Time Index")
plt.ylabel("Close Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -----------------------------
# 9) evaluation
# -----------------------------
mse = mean_squared_error(actual_close, pred_close)
mae = mean_absolute_error(actual_close, pred_close)
print(f"Evaluation — MSE: {mse:.4f}, MAE: {mae:.4f}")
```
