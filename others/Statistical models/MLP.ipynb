#MODEL1

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

class TimeSeriesPredictor(tf.keras.Model):
    def __init__(self):
        super(TimeSeriesPredictor, self).__init__()
        # لایه‌های گروهی


        self.group1 = layers.Dense(1)
        self.group2 = layers.Dense(1)

        self.fc = layers.Dense(2, activation='relu')

        # لایه خروجی نهایی (مثلاً پیش‌بینی یک مقدار سری زمانی)
        self.output_layer = layers.Dense(1)

    def call(self, inputs):
        # فرض: inputs شکل (batch_size, 310)

        x1 = inputs[:, 0:100]
        x2 = inputs[:, 100:109]


        out1 = self.group1(x1)
        out2 = self.group2(x2)


        grouped = tf.concat([out1, out2], axis=1)

        fc_out = self.fc(grouped)       # حالا fc وجود دارد
        output = self.output_layer(fc_out)

        return output
```

```python
model = TimeSeriesPredictor()

# ساخت مدل روی ورودی ساختگی برای summary
dummy_input = tf.random.normal((1, 500))
model(dummy_input)

# چاپ معماری مدل
model.summary()

# کامپایل مدل
model.compile(
    optimizer='adam',
    loss='mse',          # چون خروجی عددی است
    metrics=['mae']      # میانگین خطای مطلق هم به عنوان معیار
)
```

```python

X_train=X[0:64,:]
Y_train=y[0:64]

X_validation=X[64:96,:]
Y_validation=y[64:96]

X_test=X[96:128,:]
Y_test=y[96:128]
```

```python
print(X_train.shape)
print("*******************")
print(Y_train.shape)
print("*******************")
print(X_validation.shape)
print("*******************")
print(Y_validation.shape)
print("*******************")
print(X_test.shape)
print("*******************")
print(Y_test.shape)
```

```python
history=model.fit(X_train, Y_train, epochs=500, batch_size=4, verbose=1)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

class TimeSeriesPredictor3(tf.keras.Model):
    def __init__(self):
        super(TimeSeriesPredictor3, self).__init__()
        # لایه پنهان با 2 نورون، ورودی مستقیم از 101 ویژگی
        self.dense1 = layers.Dense(2, activation='relu')

        # لایه خروجی با 1 نورون
        self.output_layer = layers.Dense(1)

    def call(self, inputs):
        x = self.dense1(inputs)          # (None, 101) → (None, 2)
        output = self.output_layer(x)    # (None, 2) → (None, 1)
        return output
```
