```r
############################
# 1) Load packages
############################
install.packages(c("ForeCA", "astsa"))
library(ForeCA)
library(astsa)

############################
# 2) Read CSV
############################
df <- read.csv("/content/Antam.csv")

# Print column names to debug
print(colnames(df))

############################
# 3) Columns to use (ONLY these)
############################
#use_cols <- c("air_pressure","air_temp","air_wind_speed","bod","cl","cod","do","humidity","ph","temperature","toc","wt")
#use_cols <- c("Demand..MW.")
#use_cols <- c("resting_hr_bpm","spo2_avg_pct","sbp_mmHg","dbp_mmHg","caffeine_mg","stress_score")
#use_cols <- c("open","high","low","close","volume")
#use_cols <- c("AQI.Value") # Placeholder: update this after checking colnames(df)
use_cols <- c("Gold.Price")
############################
# 4) Output table
############################
results <- data.frame(
  series = character(),
  Omega = numeric(),
  stringsAsFactors = FALSE
)

############################
# 5) Compute Forecastability (Ω)
############################
for (col in use_cols) {

  x <- df[[col]]
  x <- x[!is.na(x)]   # حذف NA (اجباری برای ForeCA)

  # Skip if x is empty after NA removal
  if (length(x) == 0) {
    warning(paste("Column '", col, "' is empty or contains only NA values after filtering. Skipping.", sep=""))
    next
  }

  omega_val <- Omega(
    x,
    spectrum.control = list(method = "mvspec"),
    entropy.control  = list(prior.weight = 1e-2)
  )
omega_01 <- omega_val / 100

  results <- rbind(
    results,
    data.frame(
      series = col,
      Omega = round(omega_01,3)
    )
  )
}

############################
# 6) Show results
############################
print(results)





```
